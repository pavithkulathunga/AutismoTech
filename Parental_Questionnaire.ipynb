```python
import pandas as pd

df = pd.read_csv('Autism Dataset for Toddlers.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
columns_to_remove = ['Qchat-10-Score', 'Who completed the test', 'Case_No']  # Replace with your column names
df = df.drop(columns_to_remove, axis=1)
print(df.head())
```

```python
print(df.tail())
```

```python
df.rename(columns={'Class/ASD Traits ': 'ASD_Traits'}, inplace=True)
```

```python
print(df['Sex'].unique())
print(df['ASD_Traits'].unique())
```

```python
df['Sex'] = df['Sex'].map({'f': 0, 'm': 1})
df['Family_mem_with_ASD'] = df['Family_mem_with_ASD'].map({'yes': 1, 'no': 0})
df['ASD_Traits'] = df['ASD_Traits'].map({'Yes': 1, 'No': 0})
print(df.head())
```

```python
df.info()
```

```python
# Check the balance of each column
for column in df.columns:
    print(f"Balance for {column}:")
    print(df[column].value_counts())
    print()
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the 'ASD_Traits' column
df['ASD_Traits'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Balance of ASD_Traits (yes vs no)')
plt.xlabel('ASD Traits')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
x = df.drop("ASD_Traits", axis=1)
x.head()
```

```python
y = df["ASD_Traits"]
y.head()
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import learning_curve
```

```python
# Define models with their relevant parameters
models = {
    "Logistic Regression (C=0.1)": LogisticRegression(C=0.1),
    "K-Nearest Neighbors (n_neighbors=5)": KNeighborsClassifier(n_neighbors=5),
    "Random Forest (n_estimators=100)": RandomForestClassifier(n_estimators=100),
    "Naive Bayes": GaussianNB(),
    "SVM (C=0.1)": SVC(C=0.1),
}

# Create subplots - one for each model
fig, axes = plt.subplots(len(models), 1, figsize=(12, 8))

# Ensure axes is iterable even if there is only one subplot
if len(models) == 1:
    axes = [axes]

# Plot learning curves for each model
for ax, (model_name, model) in zip(axes, models.items()):
    # Generate learning curve
    train_sizes, train_scores, test_scores = learning_curve(
        model,
        x,
        y,
        train_sizes=np.linspace(0.1, 1.0, 10),
        cv=5,
        scoring="accuracy",
    )

    # Calculate mean and std
    train_mean = train_scores.mean(axis=1)
    train_std = train_scores.std(axis=1)
    test_mean = test_scores.mean(axis=1)
    test_std = test_scores.std(axis=1)

    # Plot learning curve
    ax.plot(train_sizes, train_mean, label=f"{model_name} - Training score", color='blue')
    ax.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.2)
    ax.plot(train_sizes, test_mean, label=f"{model_name} - Cross-validation score", color='orange')
    ax.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color='orange', alpha=0.2)

    # Title and labels
    ax.set_title(f"Learning Curve - {model_name}")
    ax.set_xlabel("Training Set Size")
    ax.set_ylabel("Accuracy")
    ax.grid(True)
    ax.legend(loc="best")

# Adjust layout
plt.tight_layout()
plt.show()
```

# **Logistic Regression**





```python
# Initialize the model
lr = LogisticRegression(C=0.1)

# Get predictions using cross-validation
y_pred = cross_val_predict(lr, x, y, cv=5)

# Calculate accuracy, precision, recall, and F1 score
accuracy = cross_val_score(lr, x, y, cv=5, scoring='accuracy').mean()
precision = cross_val_score(lr, x, y, cv=5, scoring='precision').mean()
recall = cross_val_score(lr, x, y, cv=5, scoring='recall').mean()
f1 = cross_val_score(lr, x, y, cv=5, scoring='f1').mean()

# Compute confusion matrix
cm = confusion_matrix(y, y_pred)

# Show the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(cm)
```

# **K-Nearest neighbors**

```python
# Initialize the model
knn = KNeighborsClassifier(n_neighbors=5)

# Get predictions using cross-validation
y_pred = cross_val_predict(knn, x, y, cv=5)

# Calculate accuracy, precision, recall, and F1 score
accuracy = cross_val_score(knn, x, y, cv=5, scoring='accuracy').mean()
precision = cross_val_score(knn, x, y, cv=5, scoring='precision').mean()
recall = cross_val_score(knn, x, y, cv=5, scoring='recall').mean()
f1 = cross_val_score(knn, x, y, cv=5, scoring='f1').mean()

# Compute confusion matrix
cm = confusion_matrix(y, y_pred)

# Show the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(cm)
```

# **Random Forest**

```python
# Initialize the model
rf = RandomForestClassifier(n_estimators=100)

# Get predictions using cross-validation
y_pred = cross_val_predict(rf, x, y, cv=5)

accuracy = cross_val_score(rf, x, y, cv=5, scoring='accuracy').mean()
precision = cross_val_score(rf, x, y, cv=5, scoring='precision').mean()
recall = cross_val_score(rf, x, y, cv=5, scoring='recall').mean()
f1 = cross_val_score(rf, x, y, cv=5, scoring='f1').mean()

# Compute confusion matrix
cm = confusion_matrix(y, y_pred)

# Show the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:")
print(cm)
```
